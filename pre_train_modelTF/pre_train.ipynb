```python
import tensorflow 
from tensorflow.keras.applications.inception_v3 import InceptionV3

model = InceptionV3(
    input_shape = (150,150,3),
    include_top=False,
    weights=None,
)
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
last_layer = model.get_layer('mixed10')
out_put_last = last_layer.output
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import RMSprop
```

```python
from tensorflow.keras.layers import Flatten, Dense,Dropout

x = Flatten()(out_put_last)
x = Dense(512, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(1, activation = 'sigmoid')(x)
```

```python
from tensorflow.keras.models import Model
models = Model(model.input, x)
models.summary()
```

```python

import splitfolders
splitfolders.ratio('dataset', output="output", seed=1337, ratio=(.8, 0.1,0.1)) 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train = ImageDataGenerator(rescale=1./255.,
                           rotation_range=40,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           
                           shear_range=0.2,
                           zoom_range=0.2,
                           horizontal_flip=True
                        
                           )
val = ImageDataGenerator(rescale=1./255.)

train_data = train.flow_from_directory('./output/train',
                                       batch_size=20, target_size=(150,150), class_mode='binary')


val_data = val.flow_from_directory('./output/val',
                                       batch_size=20, target_size=(150,150), class_mode='binary')
```

```python
models.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['acc'])
```

```python


history = models.fit(
    train_data,
    validation_data=val_data,   
    steps_per_epoch =100,
    epochs=5,
    validation_steps=50, verbose=1
)
```
